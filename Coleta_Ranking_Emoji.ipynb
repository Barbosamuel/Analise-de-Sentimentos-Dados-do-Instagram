```python
import pandas as pd 
import time
import os
import argparse
from selenium                       import webdriver
from selenium.webdriver.support     import expected_conditions as EC
from selenium.webdriver.support.ui  import WebDriverWait
from selenium.webdriver.common.by   import By
from webdriver_manager.chrome          import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
```

```python
# #paramentros selenium
options = webdriver.ChromeOptions()
#options.add_argument('--headless')
options.add_argument('--log-level=3')
options.add_argument('--silent')
options.add_argument('--disable-gpu')
options.add_argument("--disable-dev-shm-usage")
options.add_argument('--no-sandbox')
options.add_experimental_option('prefs', {'download.default_directory':  os.path.abspath(os.getcwd()), 'download.prompt_for_download': False, 'download.directory_upgrade': True, 'safebrowsing_for_trusted_sources_enabled': False, 'safebrowsing.enabled': False })

service_args = ['suppress-initial-diagnostic-information=true', 'hide-command-prompt-window=true', 'enable-verbose-logging=false']
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install(), service_args=service_args), options=options)
```

```python
driver.get('http://kt.ijs.si/data/Emoji_sentiment_ranking/')
```

```python
elem = driver.find_element(by=By.ID, value='myTable')
aux_dict = {'emoji':[], 'position':[],'negativo':[],'neutro':[],'positivo':[],'score':[],'unicode_name':[],'unicode_block':[]}

for tr in range(1,752):
    for td in range(1,13):
                
        if td == 1: #emoji
            
            aux_dict['emoji'].append(elem.find_element(by=By.XPATH, value='//table/tbody/tr[' + str(tr) + ']/td[' +str(td) + ']').text )
    
        elif td == 5:#position
            
             aux_dict['position'].append(elem.find_element(by=By.XPATH, value='//table/tbody/tr[' + str(tr) + ']/td[' +str(td) + ']').text )
        
        elif td == 6:#negativo
             aux_dict['negativo'].append(elem.find_element(by=By.XPATH, value='//table/tbody/tr[' + str(tr) + ']/td[' +str(td) + ']').text )
                
        elif td == 7:#neutro
             aux_dict['neutro'].append(elem.find_element(by=By.XPATH, value='//table/tbody/tr[' + str(tr) + ']/td[' +str(td) + ']').text )
                
        elif td == 8:#positivo
             aux_dict['positivo'].append(elem.find_element(by=By.XPATH, value='//table/tbody/tr[' + str(tr) + ']/td[' +str(td) + ']').text )
                
        elif td == 9:#score
             aux_dict['score'].append(elem.find_element(by=By.XPATH, value='//table/tbody/tr[' + str(tr) + ']/td[' +str(td) + ']').text )
                
        elif td == 11:#unicode name
             aux_dict['unicode_name'].append(elem.find_element(by=By.XPATH, value='//table/tbody/tr[' + str(tr) + ']/td[' +str(td) + ']').text )
                
        elif td == 12:#unicode name
             aux_dict['unicode_block'].append(elem.find_element(by=By.XPATH, value='//table/tbody/tr[' + str(tr) + ']/td[' +str(td) + ']').text )
                
    print(f'iteração {tr}')    
```

```python
df = pd.DataFrame(aux_dict)
```

```python
df['unicode'] = df['emoji'].apply(lambda x: x.encode('unicode-escape'))
```
